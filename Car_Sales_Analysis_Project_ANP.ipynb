```python
#To find the distinct Car_COmpanies in the dataset

car_companies = df["Company"].unique()
print("Car Companies:\n", car_companies)
```

```python
#Calculating total Sales
overall_sales = df['Price ($)'].sum()
print("Total sales made over the 2 years as per the dataset is :\n",overall_sales)
```

```python
#Total Sales Quantity
sales_qty = df['Car_id'].count()
print("Total number of cars sold over the period of 2 years as per the given dataset is:\n", sales_qty)
```

```python
#Average Car Price

avg_car_price = df["Price ($)"].mean().round()
print("Average car price is: \n$", avg_car_price)
```

```python
# Years
df['years'] = df['Date'].dt.year
print("The  dataset has data from the years:")
unique_years = df['years'].unique
print(df['years'].unique())
```

```python
#To find total sales for each year

total_sales = df.groupby('years').agg(total_sales = ('Price ($)', 'sum')).reset_index()
print(total_sales)
```

```python
#to show the difference in sales and percentage increment each year

sales_data = pd.DataFrame(total_sales)
sales_data['sales_increment'] = sales_data['total_sales'].diff()
sales_data['Sales_%_Increment'] = sales_data['total_sales'].pct_change()*100
print(sales_data)
```

```python
# To show Total Sales per year using column chart
x = [2022,2023]
y = [300340345, 371185120]
plt.bar(x, y, width = 0.5 )
plt.xticks(x)
plt.title("Total Sales per Year")
plt.xlabel('Years')
plt.ylabel('Total Sales')
plt.show
```

```python
#Co-relation b/w AnnualIncome of buyers and Care sales
correlation1 = df[['Annual Income', 'Price ($)']].corr()  
print(correlation1)
```

Annual Income vs. Price ($) = 0.012065: Very weak positive correlatio, meaning that as Annual Income increases, Price ($) tends to increase slightly — but the value is very close to 0, indicating that this relationship is extremely weak and likely insignificant.

```python
#Car-type Preference Comparison on the basis of Transmission
car_type_transmission = df.groupby('Transmission').agg(car_type_transmission = ('Car_id', 'count')).reset_index()
print(car_type_transmission)
```

Therefore, most Customers preferred Auto-Transmission-Cars over Manual-Transmission-Cars.

```python
transmission_pivot = pd.pivot_table(df, values = 'Car_id', index = 'Transmission', aggfunc = 'count', margins = False)
print(transmission_pivot)
```

```python
# Extract data excluding the 'All' row
transmission = transmission_pivot['Car_id']

# Plot pie chart
plt.figure(figsize=(3, 3))
transmission.plot.pie(autopct='%1.0f%%', startangle=90, colors=['#ff9999', '#66b3ff'])
plt.title('Car Preference based on Transmission Type')
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
# Sales Distribution by Car Color
car_color_pivot = pd.pivot_table(df, values = 'Price ($)', index = 'Color', aggfunc = 'sum', margins = False)
print(car_color_pivot)
```

```python
car_color_sale = car_color_pivot['Price ($)']

# Plot pie chart
plt.figure(figsize=(3, 3))
car_color_sale.plot.pie(
    autopct=lambda pct: f'{int(car_color_sale.sum() * pct / 100) / 1e6:.0f} M',  # Convert to millions with 'M'
    startangle=90,
    colors=['#ff9999', '#66b3ff', '#99ff99']
)
plt.title('Sales Distribution by Car Color')
plt.ylabel('')
plt.show()
```

*"Pale White" cars have added the maximum to the overall revenue, "Black" is next in the category and the least is added by "Red" cars.

```python
# Sales(Revenue) Distribution by Car Company
car_company_pivot = pd.pivot_table(df, values = 'Price ($)', index = 'Company', aggfunc = 'sum', margins = False)
car_company_pivot.sort_values(by='Price ($)', ascending=False, inplace= True)

print(car_company_pivot)
```

```python
# Plot horizontal bar chart
plt.figure(figsize=(8, 6))
car_company_pivot.plot.barh(color='blue', legend=False)
plt.title('Sales(Revenue) Distribution by Car Company')
plt.xlabel('Total Sales ($)')
plt.ylabel('Company')
plt.gca().invert_yaxis()  # Optional: Invert y-axis to have the top company first
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
#To show the top- 5 Car Companies which have made the most of the revenue

Top_5_sold_Companies = car_company_pivot['Price ($)'].head(5)
print(Top_5_sold_Companies)
```

```python
#To show the Bottom- 5 Car Companies which have made the least of the revenue

Bottom_5_sold_Companies = car_company_pivot['Price ($)'].tail(5)
print(Bottom_5_sold_Companies)
```

```python

df['months'] = df['Date'].dt.strftime('%b')
monthwise_sales = pd.pivot_table(df, values='Price ($)', index='months', aggfunc='sum', margins=False)
print(monthwise_sales)
```

```python
# Define the chronological order for months
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Convert the 'months' index to a Categorical type with the specified order
monthwise_sales.index = pd.Categorical(monthwise_sales.index, categories=month_order, ordered=True)

# Sort the DataFrame based on this categorical order
monthwise_sales = monthwise_sales.sort_index()

print(monthwise_sales)
```

```python
# Line chart to show Monthwise sales trend
monthwise_sales.plot(kind='line', marker='o', linestyle='-', color='blue', linewidth=2, legend = False)
plt.title('Monthwise Sales Trend')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.xticks(ticks=range(len(monthwise_sales.index)), labels=monthwise_sales.index, rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adds horizontal grid lines

plt.ticklabel_format(style='plain', axis='y')  # Disable scientific notation for y-axis

plt.show()
```

```python
monthwise_sales_yearly = pd.pivot_table(df, values='Price ($)', columns = 'years', index='months', aggfunc='sum', margins=False)
print(monthwise_sales_yearly)
```

```python
# Convert the 'months' index to a Categorical type with the specified order
monthwise_sales_yearly.index = pd.Categorical(monthwise_sales_yearly.index, categories=month_order, ordered=True)

# Sort the DataFrame based on this categorical order
monthwise_sales_yearly = monthwise_sales_yearly.sort_index()

print(monthwise_sales_yearly)
```

```python
# Line chart to show Monthwise sales trend for both the years separately
monthwise_sales_yearly.plot(kind='line', marker='o', linestyle='-', linewidth=2)
plt.title('Monthwise Sales Trend per Year')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.xticks(ticks=range(len(monthwise_sales_yearly.index)), labels=monthwise_sales_yearly.index, rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adds horizontal grid lines

plt.ticklabel_format(style='plain', axis='y')  # Disable scientific notation for y-axis

plt.show()
```

```python
# Quantities sold per Company

car_company_countpivot = pd.pivot_table(df, values = 'Car_id', index = 'Company', aggfunc = 'count', margins = False)
car_company_countpivot.sort_values(by = 'Car_id', ascending = False, inplace = True)
print(car_company_countpivot)
print("---------------------------\nTop 5\n-----------------")
print(car_company_countpivot.head(5))
print("---------------------------\nBottom 5\n-----------------")
print(car_company_countpivot.tail(5))
```

```python
# Model wise Sales performance.

# Group by Company and Model, and count the number of cars
car_model_count = df.groupby(['Company', 'Model']).size().reset_index(name='Count')

# Sort by Company first and then by count in descending order
car_model_count = car_model_count.sort_values(by=['Company', 'Count'], ascending=[True, False])

# Print the result
for company, group in car_model_count.groupby('Company'):
    print(f"\nCompany: {company}\n-------------")
    print(group[['Model', 'Count']].to_string(index=False))
```

```python
#Sales Qty by Gender

Sales_by_Gender = pd.pivot_table(df, values = 'Car_id', index = 'Gender', aggfunc = 'count', margins = False)
print(Sales_by_Gender)
```

```python
# Sale Qty by Engine Type
engine_pivot = pd.pivot_table(df, index = 'Engine', values = 'Car_id', aggfunc = 'count', margins = False).rename(columns = {'Car_id': 'Qty'})
print(engine_pivot)
```

```python
# Sale Qty by Body Style
bodystyle_pivot = pd.pivot_table(df, index = 'Body Style', values = 'Car_id', aggfunc = 'count', margins = False).rename(columns = {'Car_id': 'Qty'})
bodystyle_pivot = bodystyle_pivot.sort_values(by = 'Qty', ascending = False)
print(bodystyle_pivot)
```

```python
# Region-wise Sales each year
#creating a pivot table with Regions as Row Indices and Sum of sales made as values
Regionwise_pivot = pd.pivot_table(df, columns='years', index='Dealer_Region', values='Price ($)', aggfunc='sum', margins=False)

# Rename the column level from 'Price ($)' to 'Sales'
Regionwise_pivot.columns.name = 'Sales'

# Add a new column 'Total Revenue' with the sum of sales across all years for each region
Regionwise_pivot['Total Revenue'] = Regionwise_pivot.sum(axis=1)

# Sort the pivot table by the total sales across all years in descending order (simplified)
Regionwise_pivot = Regionwise_pivot.loc[Regionwise_pivot.sum(axis=1).sort_values(ascending=False).index]

# Print the sorted pivot table
print(Regionwise_pivot)
```
